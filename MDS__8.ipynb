### Задание 2


Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

### Решение:

1. События называются попарно независимыми, если для любой пары событий выполняется: $P(A_i A_j) = P(A_i)P(A_j)$

$ m1 = |O| |P| ; m2 = |O| |P| ; m3 = |O| |P| $  

Вероятность выпадения каждой стороны $0.5$

$ \Omega = |O , O| |O, P| |P , O| |P, P| $

Вероятность каждого из исходов $0.25$

$A1 = |O , O| |P, P|; $

$A2 = |O, O | |P, P|; $

$A3 = |O, O | |P, P|$

```python
# Пусть А1 = m1 и m2 выпал орел:
0.5 * 0.5 == 0.25
```

```python
# Пусть А2 = m2 и m3 выпал орел:
0.5 * 0.5 == 0.25
```

```python
# Пусть А3 = m2 и m3 выпал орел:
0.5 * 0.5 == 0.25
```

Ч.т.д. : события являются попарно независимыми

2. События называются независимыми в совокупности, если любое из них не зависит от любого другого события и от любой комбинации остальных
$P(A_1 A_2 ... A_n) = P(A_1)P(A_2)...P(A_n)$

$ \Omega = |O,O,O| |O,O,P| |O,P,O| |O,P,P|$ 
$|P,O,O| |P,O,P| |P,P,O| |P,P,P|$

Вероятность каждого из исходов $0.125$

```python
# Пусть на всех 3-ех монетах выпал орел -> произошли события А1, А2 и А3 одновременно
0.125 == 0.25 * 0.25 * 0.25
```

Ч.т.д. : события не являются независимыми в совокупности.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

#### 1. Нормальное распределение

**scipy.stats.norm.pdf(data, loc, scale)** 

data: это набор точек или значений, которые представляют равномерно отобранные данные в виде данных массива

loc: используется для указания среднего значения, по умолчанию равно 0

scale: используется для указания стандартного отклонения, по умолчанию это 1


```python
observatin_x = np.linspace(-4,4,200)
pdf_norm = stats.norm.pdf(observatin_x, loc=0, scale=1)
# pdf_norm
```

```python
def norm_graf():
    plt.plot(observatin_x, pdf_norm)
    plt.xlabel('x-values')
    plt.ylabel('pdf_norm_values')
    plt.title("Probability density funciton of normal distribution")
    plt.show()
```

```python
norm_graf()
```

```python
# 1) Увеличение интервала исходных значений, при неизменных среднем значении и среднеквадратичном отклонении 
# создает большие "хвосты" с околонулевой вероятностью

observatin_x = np.linspace(-10,10,200)
pdf_norm = stats.norm.pdf(observatin_x, loc=0, scale=1)
norm_graf()
```

```python
# 2) Увеличение/уменьшение среднего значения (loc) смещает вершину распределения вправо/влево соответственно 
observatin_x = np.linspace(-4,4,200)
pdf_norm = stats.norm.pdf(observatin_x, loc=2, scale=1)
norm_graf()
```

```python
# 3) Если мат. ожидание = среденее значение (loc) выходит за границы растперделения, график перестает быть информативным
observatin_x = np.linspace(-4,4,200)
pdf_norm = stats.norm.pdf(observatin_x, loc=10, scale=1)
norm_graf()
```

```python
# 4) Среднеквадратичное отклонение (default = 1) влияет на ширину "колокола" - дисперсию
# scale не может = 0

observatin_x = np.linspace(-4,4,200)
pdf_norm = stats.norm.pdf(observatin_x, loc=0, scale=0.5)
norm_graf()
```

```python
# scale = (0, +oo)

observatin_x = np.linspace(-4, 4, 200)
pdf_norm = stats.norm.pdf(observatin_x, loc=0, scale=5)
norm_graf()
```

```python
def expon_graf():
    plt.plot(array_data, expon_pdf_data, 'c.') 
    plt.xlabel('x-values')
    plt.ylabel('expon_pdf_data')
    plt.title("Probability density funciton of expotential distribution")
    plt.show()
```

#### 3. Распределение Стьюдента

```python
from scipy.stats import t
```

t.ppf(<точность>, степень свободы)

```python
df = 4
x = np.linspace(t.ppf(0.01, df), t.ppf(0.99, df), 100)
```

```python
def t_graf():
    fig, ax = plt.subplots(1, 1)
    ax.plot(x, t.pdf(x, df), 'm-')
    plt.xlabel('x-values')
    plt.ylabel('pdf_norm_values')
    plt.title("Probability density funciton of Student’s t continuous distribution")
    plt.show()
```

```python
t_graf()
```

Выглядит аналогично графику плотности нормального распределения, но уже и ниже

```python
# 1) Чем выше степень свободы, тем "шире" купол = больше дисперсия
df = 20
x = np.linspace(t.ppf(0.01, df), t.ppf(0.99, df), 100)
t_graf()
```

```python
# 2) Если точность не дает в сумме 1, то график получается несимметриным
df = 2.74
x = np.linspace(t.ppf(0.1, df), t.ppf(0.8, df), 100)
t_graf()
```

```python
# 3) Отрицательная ошибка и ошибка > 1 не существуют
df = 2
x = np.linspace(t.ppf(-0.05, df), t.ppf(1.5, df), 100)
t_graf()
```

```python
# 4) чем ниже точность, тем "шире" колокол = дисперсия больше
df = 2
x = np.linspace(t.ppf(0.2, df), t.ppf(0.8, df), 100)
t_graf()
```
