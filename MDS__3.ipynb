## Уровень 0:

### Задание 1
Дан объект в *2D* пространстве

```python
A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
```

При помощи линейных отображений:

- Уменьшить объект в два раза



```python
# можно умножить каждую на 0.5, используя диагональную матрицу 3х3
B = np.array([[0.5,  0, 0], [0, 0.5, 0], [0, 0, 0.5]])
B
```

```python
A1 = np.array(A @ B)

x = A1[:,0]
y = A1[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
```

- Повернуть на 130 градусов


```python
# используем ортогональную матрицу
rad = math.radians(130)
R = np.array([[np.cos(rad), -np.sin(rad), 0], 
               [np.sin(rad), np.cos(rad), 0], 
               [0, 0, 1]])
R
```

```python
A2 = np.array(A @ R)

x = A2[:,0]
y = A2[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
```

- Отразить объект относительно прямой y = x 

поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]] 

```python
C = np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])
C
```

```python
A3 = np.array(A @ C)
x = A3[:,0]
y = A3[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

### Задание 2
Найдите спектральное разложение матрицы:

M = S * D * St
- S - матрица собственных векторов
- D - диагональная матрица собственных значений
- St - транспонированая S

```python
M = np.array([[1, 2],
              [2, 3]])
M
```

```python
# собственные значения и векторы
d, S = np.linalg.eig(M)
D = np.diag(d)
S
```

```python
D
```

```python
St = np.transpose(S)
St
```

```python
result = S @ D @ St
result
```
