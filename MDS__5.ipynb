### Задание 1

Посчитайте частные производные функций:

#### 1) $f(x,y)=2x^2y^3 + 1/x + y^2x + 7$


```python
x = sym.Symbol('x')
y = sym.Symbol('y')

# частная производная первого порядка по х
f1 = 2*x**2 * y**3 + 1/x + y**2 * x + 7
sym.diff(f1, x)
```

```python
# частная производная первого порядка по y
sym.diff(f1, y)
```

```python
# частная производная второго порядка по х
fx1 = 4*x*y**3 + y**2 - 1/x**2
sym.diff(fx1, x)
```

```python
# частная производная второго порядка по y
fy1 = 6*x**2*y**2 + 2*x*y
sym.diff(fy1, y)
```

```python
# смешанная частная производная второго порядка  
fxy1 = 4*x*y**3 + y**2 - 1/x**2
sym.diff(fx1, y)
```

#### 2) $f(x,y)=x^2y - sin(xy) + cos(x^2) + 6y$

```python
x = sym.Symbol('x')
y = sym.Symbol('y')

f2 = x**2 * y - sym.sin(x*y) + sym.cos(x**2) + 6*y
f2
```

```python
# частная производная первого порядка по х
sym.diff(f2, x)
```

```python
# частная производная первого порядка по y
sym.diff(f2, y)
```

```python
# частная производная второго порядка по х
fx2 = 2*x*y - 2*x*sym.sin(x**2) - y*sym.cos(x*y)
sym.diff(fx2, x)
```

```python
# частная производная второго порядка по y
fy2 = x**2 - x*sym.cos(x*y) + 6
sym.diff(fy2, y)
```

```python
# смешанная частная производная второго порядка  
fyx2 = x**2 - x*sym.cos(x*y) + 6
sym.diff(fyx2, x)
```

Постройте трехмерный график функции

$f(x)=x_2^2-x_1^2 x_2 + 2 x_1 x_2$

```python
x = sym.Symbol('x')
y = sym.Symbol('y')

f = y**2 - x**2 * y + 2 * x * y
f
```

```python
from mpl_toolkits import mplot3d

x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)
z = y**2 - x**2 * y + 2 * x * y
 
fig = plt.figure()
 
# syntax for 3-D plotting
ax = fig.add_subplot(111, projection='3d')
 
# syntax for plotting
ax.plot_surface(x, y, z, edgecolor ='pink')
ax.set_title('My graf')
plt.show()
```
