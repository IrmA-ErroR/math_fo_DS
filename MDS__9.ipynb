### Распределение Парето

```python
from scipy.stats import pareto
b = 20
rv = pareto(b)
sample = rv.rvs(size=1000)
sample[:20]
```

```python
# визуализация распределения Парето
x = np.linspace(pareto.ppf(0.01, b), pareto.ppf(0.99, b), 100)
plt.plot(x, pareto.pdf(x, b), '-', lw=5, alpha=0.8, label='pareto pdf')
plt.show()
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples f(x)')
plt.xlabel('x')
plt.show()
```
